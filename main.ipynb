### Visualizations for Analysis

#### **I. Correlation Plots**

```python
cmap = plt.cm.magma
plt.figure(figsize=(18, 16))
plt.title('Pearson Correlation of Features', y=1.02, size=18)
sns.heatmap(voice_data.iloc[:, :-1].astype(float).corr(), linewidth = 0.5, vmax = 1, square = True, cmap=cmap, linecolor='white', annot=True)
```

#### **II. Implementing TNSE**

```python
X_data = voice_data.iloc[:, :-1]
Y_data = voice_data['label']
Y_data = Y_data.replace({'male':1, 'female':0})
```

```python
X_scale = StandardScaler().fit_transform(X_data)
tsne = TSNE(n_components=2, init='pca', random_state=0)
X_t = tsne.fit_transform(X_scale)
```

```python
plt.figure()
plt.scatter(X_t[np.where(Y_data == 0), 0], X_t[np.where(Y_data == 0), 1], marker='x', color='yellow', linewidth=1.2, alpha = 0.5, label='Male')
plt.scatter(X_t[np.where(Y_data == 1), 0], X_t[np.where(Y_data == 1), 1], marker='v', color='pink', linewidth=1.2, alpha = 0.5, label='Male')

plt.xlabel('Dimension I')
plt.ylabel('Dimension II')
plt.title('T-SNE')
plt.legend(loc='best')
plt.show()
```
